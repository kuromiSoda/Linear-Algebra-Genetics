```python
import numpy as np
```

Stochastic Matrix obtained from AA-AA and AA-Aa

```python
M = np.array([[1.0, 0.5], [0.0, 0.5]])
```

Initial generation

```python
x0 = np.array([[0.9],[0.1]])
```

Input for number of gens

```python
n = input("Enter the number of years after which the result is needed: ")
n = int(n)
```

Diagonalisation of matrix

```python
evals, evects = np.linalg.eig(M)

P = evects

D = np.zeros((2,2))
for i in range(2):
    D[i,i] = evals[i]

P_inv = np.linalg.inv(P)
```

```python
Dn = np.linalg.matrix_power(D,n)
```

The result

```python
xn = np.dot((P @ Dn @ P_inv), x0)
xn
```
